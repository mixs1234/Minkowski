```python
import ipywidgets as widgets
from matplotlib.figure import Figure
import matplotlib.pyplot as a
import numpy as np
import matplotlib

n = np.linspace(0, 10, 100)
#indsæt din egen variable mellem x=[0,8] og ct=[0,8]
Variable = [3, 4]


def gamma(beta_):
    return (1 - beta_ ** 2) ** (-1 / 2)


def x_mærke(num):
    return [2 * np.cos(np.arctan(float(num))), 2 * np.sin(np.arctan(float(num)))]


def ct_mærke(num):
    return [2 * np.sin(np.arctan(float(num))), 2 * np.cos(np.arctan(float(num)))]


def unitvektor_x(beta, gamma):
    return [gamma, gamma * beta]


def unitvektor_ct(beta, gamma):
    return [gamma * beta, gamma]


def position(beta):
    ct_m = gamma(beta) * (Variable[1] - beta * Variable[0])
    x_m = gamma(beta) * (Variable[0] - beta * Variable[1])
    return [gamma(beta) * x_m, gamma(beta) * (beta * x_m), gamma(beta) * beta * ct_m, gamma(beta) * ct_m]


def lorenz_trans(beta=0):
    a.plot(n, n, color='yellow', label='Lys')
    a.arrow(0, 0, 8, 0, width=0.02, color='blue', label='x, ct')
    a.arrow(0, 0, 0, 8, width=0.02, color='blue')
    a.arrow(0, 0, 2 * 3 * 2 * x_mærke(beta)[0], 2 * 3 * 2 * x_mærke(beta)[1], width=0.02, color='red', label='x\',ct\'')
    a.arrow(0, 0, 2 * 3 * 2 * ct_mærke(beta)[0], 2 * 3 * 2 * ct_mærke(beta)[1], width=0.02, color='red')
    a.ylim([-0.2, 8])
    a.xlim([-0.2, 8])
    # Grid
    for r in range(9):
        a.arrow(r * gamma(beta), r * gamma(beta) * beta, 10 * gamma(beta) * beta, 10 * gamma(beta), linestyle='--',
                color='gray', width=0.001)
        a.arrow(r * gamma(beta) * beta, r * gamma(beta), 10 * gamma(beta), 10 * gamma(beta) * beta, linestyle='--',
                color='gray', width=0.001)

    a.plot(Variable[0], Variable[1], marker="*", markerfacecolor="red", markersize=8)

    # Event
    a.plot([0, Variable[0]], [Variable[1], Variable[1]], linestyle="dashed", color="blue")
    a.plot([Variable[0], Variable[0]], [0, Variable[1]], linestyle="dashed", color="blue")

    # Event-mærkedesystem
    a.plot([position(beta)[0], Variable[0]], [position(beta)[1], Variable[1]], linestyle="dashed", color="red")
    a.plot([position(beta)[2], Variable[0]], [position(beta)[3], Variable[1]], linestyle="dashed", color="red")
    a.title("Minkowski diagram")
    a.legend()
    a.grid()
    a.show()


widgets.interact(lorenz_trans, beta=(0, 1, 0.01));
```
